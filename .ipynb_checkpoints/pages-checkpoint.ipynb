```python
import pygame
from ipynb.fs.full.utils import util

import pygame
import pickle

        
class home:
    
    def __init__(self, surface, width, rows, pageManager):
        self.surface = surface
        self.width = width
        self.rows = rows
        self.running = True
        self.clock = pygame.time.Clock()
        self.exit = None
        self.pageManager = pageManager
        self.toRender = []
        self.toReturn = None
            
    def renderHomeScreen(self):
        outer = pygame.Surface((self.width, self.width))
        
        background = (pygame.image.load('images/gameStartBackground.png').convert_alpha())
        pygame.transform.scale(background, (self.width,self.width))
        outer.blit(background,(0,0))
        filler = pygame.Surface((self.width, self.width))
        filler.fill((255,255,255, 128)) 
        
        outer.blit(filler, (0,0))
        
        heeia = (pygame.image.load('images/loko/Heeia.png').convert_alpha())
        heeia = pygame.transform.scale(heeia, (int(116*1.2),int(118*1.2)))
        heeia = util.makeOutline(heeia, 3)
        
        sizes = heeia.get_size()
        
        tray = pygame.Surface((sizes[0], sizes[1]+30))
        tray.blit(heeia, (0,0))
        tray.blit(util.makeText("He`eia Loko", 30), (sizes[0]//2 - util.makeText("He`eia Loko", 30).get_width()//2, sizes[1]))
        heeia = tray
        
        waikalua = (pygame.image.load('images/loko/Waikalua.png').convert_alpha())
        waikalua =pygame.transform.scale(waikalua, (int(116*1.2),int(118*1.2)))
        waikalua = util.makeOutline(waikalua, 3)
        tray = pygame.Surface((sizes[0], sizes[1]+30))
        tray.blit(waikalua, (0,0))
        tray.blit(util.makeText("Waikalua Loko", 30), (sizes[0]//2 - util.makeText("Waikalua Loko", 30).get_width()//2, sizes[1]))
        waikalua = tray
        
        self.heeiaButton = outer.blit(heeia, (self.width//2 - sizes[0]//2 - sizes[0] ,self.width//2))
        self.waikaluaButton = outer.blit(waikalua, (self.width//2 + sizes[0]//2,self.width//2))
        
        title = util.makeText("Da Loko Game!", 64)
        outer.blit(title, (self.width//2 - title.get_width()//2, 100))
        
        self.surface.blit(outer, (0,0))
        
    def handleEvent(self, i):
        pos = pygame.mouse.get_pos()
        if i.type == pygame.MOUSEBUTTONUP:
            if self.heeiaButton.collidepoint(pos):
                self.running = False
                self.toReturn = self.pageManager["Heeia"]
            if self.waikaluaButton.collidepoint(pos):
                self.running = False
                self.toReturn = self.pageManager["Waikalua"]

    def update(self):
        pos = pygame.mouse.get_pos()
        if self.heeiaButton.collidepoint(pos):
            heeiaTray = pygame.Surface((150,20))
            heeiaTray.fill((225, 230, 226, 180))
            sizes = heeiaTray.get_size()
            heeiaText = util.makeText("Click to go to He`eia!", 20)
            heeiaTray.blit(heeiaText, (sizes[0]//2 - heeiaText.get_width()//2,5))
            heeiaTray = util.makeOutline(heeiaTray, 2)
            self.toRender.append({"elem": heeiaTray, "pos": (pos[0] - 50, pos[1] - 20)})

        if self.waikaluaButton.collidepoint(pos):
            waikaluaTray = pygame.Surface((150,20))
            waikaluaTray.fill((225, 230, 226, 180))
            sizes = waikaluaTray.get_size()
            waikaluaText = util.makeText("Click to go to Waikalua!", 20)
            waikaluaTray.blit(waikaluaText, (sizes[0]//2 - waikaluaText.get_width()//2,5))
            waikaluaTray = util.makeOutline(waikaluaTray, 2)
            self.toRender.append({"elem": waikaluaTray, "pos": (pos[0] - 50, pos[1] - 20)})
    
    def reDraw(self):
        self.renderHomeScreen()
        while len(self.toRender) != 0:
            i = self.toRender.pop()
            self.surface.blit(i["elem"], i["pos"])
            
    def mainLoop(self):
        self.renderHomeScreen()

        while self.running:

            # --- events ---
            
            for event in pygame.event.get():

                # --- global events ---
                
                if event.type == pygame.QUIT:
                    self.is_running = False
                elif event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_ESCAPE:
                        self.is_running = False

                # --- objects events ---
#                     '''
#                     for widget in self.widgets:
#                         widget.handle_event(event)
#                     '''
                self.handleEvent(event)
                
            # --- updates ---
#                 '''
#                 for widget in self.widgets:
#                     widget.update()
#                 '''
            self.update()
            
            
            # --- draws ---
#                 '''
#                 for widget in self.widgets:
#                     widget.draw(surface)
#                 '''
            self.reDraw()
            
            pygame.display.update()

            # --- FPS ---

            self.clock.tick(25)

        # --- the end ---
    
        return self.toReturn

class Heeia:
    
    def __init__(self, surface, width, rows, pageManager):
        self.surface = surface
        self.width = width
        self.rows = rows
        self.running = True
        self.clock = pygame.time.Clock()
        self.exit = None
        self.pageManager = pageManager
        self.toRender = []
        self.toReturn = None
            
    def renderHomeScreen(self):
        outer = pygame.Surface((self.width, self.width))
        inner = pygame.Surface((self.width, int(self.width*.7)))
        inner.fill((255,255,255,150))
        background = (pygame.image.load('images/gameStartBackground.png').convert_alpha())
        pygame.transform.scale(background, (self.width,self.width))
        outer.blit(background,(0,0))
        filler = pygame.Surface((self.width, self.width))
        filler.fill((255,255,255, 128)) 
        
        outer.blit(filler, (0,0))
        outer.blit(inner, (0,int(self.width*.3//2) + 20))
        
        heeia = (pygame.image.load('images/loko/Heeia.png').convert_alpha())
        heeia = pygame.transform.scale(heeia, (int(116*3),int(118*3)))
        heeia = util.makeOutline(heeia, 3)
        
        sizes = heeia.get_size()
        
        heeia = outer.blit(heeia, (self.width - sizes[0]//4 - sizes[0] ,self.width//2))
        
        title = util.makeText("E Komo Mai o He`eia", 64)
        titleBlock = outer.blit(title, (100, 200))
        
        description = []
        descSize = 30
        description.append(util.makeText("He'eia fishpond is Hawaiian fishpond located in He'eia uli, Ko'olaupoko, O'ahu.", descSize))
        description.append(util.makeText("It consists of 88 acres of brackish water. He'eia fishpond is built on the ", descSize))
        description.append(util.makeText("Malaukua'a fringing reef that extends from the shoreline surrounding the pond", descSize))
        description.append(util.makeText("out into Kane'ohe Bay. It was built 600-800 years", descSize))
        description.append(util.makeText("ago by the inhabitants of that area.", descSize))
        
        for c, i in enumerate(description):
            lowest = outer.blit(i, (100, titleBlock.bottom + (c+1) * 20))
        
        objTitle = util.makeText("In this level, you will", 50)
        lowest = outer.blit(objTitle, (100, lowest.bottom + 60 ))
        objTitle = util.makeText("learn about:", 50)
        lowest = outer.blit(objTitle, (100, lowest.bottom + 10 ))
        
        objectives = []
        objectives.append(util.makeText("Reef Types", descSize))
        objectives.append(util.makeText("Native and Invasive species", descSize))
        objectives.append(util.makeText("Dissolved Oxygen", descSize))
        objectives.append(util.makeText("Mo`olelo", descSize))
        
        for c, i in enumerate(objectives):
            bot = outer.blit(i, (150, heeia.top + (c + 1)  * 30))
            
        nextBut = (pygame.image.load('images/next.png').convert_alpha())
        homeBut = (pygame.image.load('images/home.png').convert_alpha())
        nextBut = pygame.transform.scale(nextBut, (int(145*.8), int(58*.8)))
        self.nextButton = outer.blit(nextBut, (150, heeia.bottom - 47))
        self.homeBut = outer.blit(homeBut, (50,50))
        
        self.surface.blit(outer, (0,0))
        
    def handleEvent(self, i):
        pos = pygame.mouse.get_pos()
        if i.type == pygame.MOUSEBUTTONUP:
            if self.nextButton.collidepoint(pos):
                self.is_running = False
                self.toReturn = "HeeiaLoko"
            if self.homeBut.collidepoint(pos):
                self.is_running = False
                self.toReturn = self.pageManager["Home"]

    def update(self):
#         pos = pygame.mouse.get_pos()
#         if self.heeiaButton.collidepoint(pos):
#             heeiaTray = pygame.Surface((150,20))
#             heeiaTray.fill((225, 230, 226, 180))
#             sizes = heeiaTray.get_size()
#             heeiaText = util.makeText("Click to go to He`eia!", 20)
#             heeiaTray.blit(heeiaText, (sizes[0]//2 - heeiaText.get_width()//2,5))
#             heeiaTray = util.makeOutline(heeiaTray, 2)
#             self.toRender.append({"elem": heeiaTray, "pos": (pos[0] - 50, pos[1] - 20)})

#         if self.waikaluaButton.collidepoint(pos):
#             waikaluaTray = pygame.Surface((150,20))
#             waikaluaTray.fill((225, 230, 226, 180))
#             sizes = waikaluaTray.get_size()
#             waikaluaText = util.makeText("Click to go to Waikalua!", 20)
#             waikaluaTray.blit(waikaluaText, (sizes[0]//2 - waikaluaText.get_width()//2,5))
#             waikaluaTray = util.makeOutline(waikaluaTray, 2)
#             self.toRender.append({"elem": waikaluaTray, "pos": (pos[0] - 50, pos[1] - 20)})
        pass

    def reDraw(self):
        self.renderHomeScreen()
#         while len(self.toRender) != 0:
#             i = self.toRender.pop()
#             self.surface.blit(i["elem"], i["pos"])

    def mainLoop(self):

        self.is_running = True

        while self.is_running:
            for event in pygame.event.get():
                # --- global events ---
                if event.type == pygame.QUIT:
                    self.is_running = False
                elif event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_ESCAPE:
                        self.is_running = False
                # --- objects events ---
                self.handleEvent(event)
            # --- updates ---
            self.update()
            # --- draws ---
            self.reDraw()

            pygame.display.update()
            # --- FPS ---
            self.clock.tick(25)
    
        return self.toReturn
    
class Waikalua:
    
    def __init__(self, surface, width, rows, pageManager):
        self.surface = surface
        self.width = width
        self.rows = rows
        self.running = True
        self.clock = pygame.time.Clock()
        self.exit = None
        self.pageManager = pageManager
        self.toRender = []
        self.toReturn = None
            
    def renderHomeScreen(self):
        outer = pygame.Surface((self.width, self.width))
        inner = pygame.Surface((self.width, int(self.width*.7)))
        inner.fill((255,255,255,150))
        background = (pygame.image.load('images/gameStartBackground.png').convert_alpha())
        pygame.transform.scale(background, (self.width,self.width))
        outer.blit(background,(0,0))
        filler = pygame.Surface((self.width, self.width))
        filler.fill((255,255,255, 128)) 
        
        outer.blit(filler, (0,0))
        outer.blit(inner, (0,int(self.width*.3//2) + 20))
        
        heeia = (pygame.image.load('images/loko/Waikalua.png').convert_alpha())
        heeia = pygame.transform.scale(heeia, (int(116*3),int(118*3)))
        heeia = util.makeOutline(heeia, 3)
        
        sizes = heeia.get_size()
        
        heeia = outer.blit(heeia, (self.width - sizes[0]//4 - sizes[0] ,self.width//2))
        
        title = util.makeText("E Komo Mai o Waikalua", 64)
        titleBlock = outer.blit(title, (100, 200))
        
        description = []
        descSize = 30
        description.append(util.makeText("The Waikalua Loko fishpond is located in Kāne‘ohe Bay and is approximately 400-year-old.", descSize))
        description.append(util.makeText("The purpose of the loko is a means to supply protein to the Kaneohe Ahupua'a on a sustainable basis. ", descSize))
        description.append(util.makeText("Originally it was a single set wall. In the past 8 yrs, another side of the wall has been added", descSize))
        description.append(util.makeText("the materials of the wall are pohaku and coral.", descSize))
        description.append(util.makeText("Currently Waikalua Loko is growing native limu such as manuea, palahalaha,", descSize))
        description.append(util.makeText("lepe o hina, and ele ele in tanks and directly in the pond.", descSize))
        
        for c, i in enumerate(description):
            lowest = outer.blit(i, (100, titleBlock.bottom + (c+1) * 20))
        
        objTitle = util.makeText("In this level, you will", 50)
        lowest = outer.blit(objTitle, (100, lowest.bottom + 60 ))
        objTitle = util.makeText("learn about:", 50)
        lowest = outer.blit(objTitle, (100, lowest.bottom + 10 ))
        
        objectives = []
        objectives.append(util.makeText("Fish tank systems", descSize))
        objectives.append(util.makeText("Native and Invasive species", descSize))
        objectives.append(util.makeText("Tide and turbidity", descSize))
        objectives.append(util.makeText("Mo`olelo", descSize))
        
        for c, i in enumerate(objectives):
            bot = outer.blit(i, (150, heeia.top + (c + 1)  * 30))
            
        nextBut = (pygame.image.load('images/next.png').convert_alpha())
        homeBut = (pygame.image.load('images/home.png').convert_alpha())
        nextBut = pygame.transform.scale(nextBut, (int(145*.8), int(58*.8)))
        self.nextButton = outer.blit(nextBut, (150, heeia.bottom - 47))
        self.homeBut = outer.blit(homeBut, (50,50))
        
        self.surface.blit(outer, (0,0))
        
    def handleEvent(self, i):
        pos = pygame.mouse.get_pos()
        if i.type == pygame.MOUSEBUTTONUP:
            if self.nextButton.collidepoint(pos):
                self.is_running = False
                self.toReturn = "WaikaluaLoko"
            if self.homeBut.collidepoint(pos):
                self.is_running = False
                self.toReturn = self.pageManager["Home"]

    def update(self):
#         pos = pygame.mouse.get_pos()
#         if self.heeiaButton.collidepoint(pos):
#             heeiaTray = pygame.Surface((150,20))
#             heeiaTray.fill((225, 230, 226, 180))
#             sizes = heeiaTray.get_size()
#             heeiaText = util.makeText("Click to go to He`eia!", 20)
#             heeiaTray.blit(heeiaText, (sizes[0]//2 - heeiaText.get_width()//2,5))
#             heeiaTray = util.makeOutline(heeiaTray, 2)
#             self.toRender.append({"elem": heeiaTray, "pos": (pos[0] - 50, pos[1] - 20)})

#         if self.waikaluaButton.collidepoint(pos):
#             waikaluaTray = pygame.Surface((150,20))
#             waikaluaTray.fill((225, 230, 226, 180))
#             sizes = waikaluaTray.get_size()
#             waikaluaText = util.makeText("Click to go to Waikalua!", 20)
#             waikaluaTray.blit(waikaluaText, (sizes[0]//2 - waikaluaText.get_width()//2,5))
#             waikaluaTray = util.makeOutline(waikaluaTray, 2)
#             self.toRender.append({"elem": waikaluaTray, "pos": (pos[0] - 50, pos[1] - 20)})
        pass

    def reDraw(self):
        self.renderHomeScreen()
#         while len(self.toRender) != 0:
#             i = self.toRender.pop()
#             self.surface.blit(i["elem"], i["pos"])

    def mainLoop(self):

        self.is_running = True

        while self.is_running:
            for event in pygame.event.get():
                # --- global events ---
                if event.type == pygame.QUIT:
                    self.is_running = False
                elif event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_ESCAPE:
                        self.is_running = False
                # --- objects events ---
                self.handleEvent(event)
            # --- updates ---
            self.update()
            # --- draws ---
            self.reDraw()

            pygame.display.update()
            # --- FPS ---
            self.clock.tick(25)
    
        return self.toReturn
```
