```python
import pygame
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
import matplotlib.backends.backend_agg as agg
import pylab


class sensorDash:
    
    def __init__(self, sensors, width, surface):
        self.showing = False
        self.typesShowing = []
        self.sensors = sensors
        self.width = width
        self.surface = surface
        self.typeSelector = []
        
    def update(self):
#         if self.showing:
#             self.draw()
        pass
    
    def draw(self):
        types = self.sensors[0].kinds

        
#         draw tray
#         add title
#         for i in self.typesShowing:
#         draw each little screen

#       Draw the check boxes to show types
        width = self.width
        outer = pygame.Surface((width, width))
        outer.fill((255,255,255, 128)) 
        tray1 = pygame.Surface((width - 300,width - 300))
        tray1.fill((255,255,255)) 
        font = pygame.font.Font('freesansbold.ttf', 20) 
        title = font.render("Data Dashboard", True, (0,0,0)) 
        tray1.blit(title, (25,50))    
        pygame.draw.line(tray1, (0,0,0), (0,0), (width - 300,0), 3)
        pygame.draw.line(tray1, (0,0,0), (0,0), (0,width - 300), 3)
        pygame.draw.line(tray1, (0,0,0), (0,width - 300), (width - 300,width - 300), 3)
        pygame.draw.line(tray1, (0,0,0), (width - 300,0), (width - 300,width - 300), 3)    
        box = pygame.image.load('graphicsPM/closeButton.png').convert_alpha()
        box = pygame.transform.scale(box, (15,15))
        
        close = tray1.blit(box, (width - 300 - 20, 10))        

        w = 180
        h = 300
        tray = pygame.Surface((w,h))
        tray.fill((225, 230, 226)) 
        font = pygame.font.Font('freesansbold.ttf', 15) 
        title = font.render("Show data points", True, (0,0,0)) 
        tray.blit(title, (5,5))    
        font2 = pygame.font.Font('freesansbold.ttf', 12) 
        self.typeSelctor = []
        for c, i in enumerate(types):
            name = i
            isShowing = name in self.typesShowing
            
            pressable = pygame.Surface((w, 15))
            pressable.blit(font2.render(name, True, (0,0,0)), (15,0))
            if not isShowing:
                box = pygame.image.load('graphicsPM/emptyBox.png').convert_alpha()
                box = pygame.transform.scale(box, (10,10))
            else:
                box = pygame.image.load('graphicsPM/checkedBox.png').convert_alpha()
                box = pygame.transform.scale(box, (10,10))
            toggleBut = pressable.blit(box, (2, 0))
            self.typeSelector.append({i: tray.blit(pressable, (10, 40 + c * 15))})
        pygame.draw.line(tray, (0,0,0), (0,0), (w,0), 3)
        pygame.draw.line(tray, (0,0,0), (0,0), (0,h), 3)
        pygame.draw.line(tray, (0,0,0), (0,h), (w,h), 3)
        pygame.draw.line(tray, (0,0,0), (w,0), (w,h), 3)    
        
        tray1.blit(tray, (width - 300 - 200, 75))
        
        tray2 = pygame.Surface((width - 100 - w - 20, width - 70))
        for c, i in enumerate(self.typesShowing):
            dat = []
            toDo = len(self.sensors[0].data[i])
            if toDo > 650:
                toDo = 650
            for j in range(toDo): #the oldest sensors kind data

                row = []
                for l in self.sensors:
                    data = l.data[i] #data from sensor l of type i
                    if len(data) > j:
                        elem = data.copy()#Make everyones most recent entry show first
                        elem.reverse()
                        row.append(elem[j])
                entry = sum(row) / len(row)
                dat.append(entry)     

            fig = pylab.figure(figsize=[3.75, 3.75], # Inches
                               dpi=100,        # 100 dots per inch, so the resulting buffer is 400x400 pixels
                               )
            ax = fig.gca()
            dat.reverse()
            ax.plot(dat)
            ax.set_title("Average {} from sensors".format(i))
            ax.set_xlabel('Time')
            ax.set_ylabel(i)
            
            canvas = agg.FigureCanvasAgg(fig)
            canvas.draw()

            renderer = canvas.get_renderer()
            raw_data = renderer.tostring_rgb()

            size = canvas.get_width_height()

            toPlot = pygame.image.fromstring(raw_data, size, "RGB")
            tray2.blit(toPlot, (10, 75 + 0*10))
            tray1.blit(tray2, (10, 5))

        outer.blit(tray1, ((width/2 - (width - 300)/2), (width/2 - (width - 300)/2)))
    #             self.surface.blit(outer, (width - 100 / 2, width - 100 / 2))
        return [outer, close, self.typeSelector]
        
    def drawButton(self): 
        w = 180
        h = 20
        tray = pygame.Surface((w,h))
        tray.fill((255,255,255)) 
        font = pygame.font.Font('freesansbold.ttf', 15) 
        title = font.render("Open Data Dashboard", True, (0,0,0)) 
        tray.blit(title, (5,5))
        pygame.draw.line(tray, (0,0,0), (0,0), (w,0), 3)
        pygame.draw.line(tray, (0,0,0), (0,0), (0,h), 3)
        pygame.draw.line(tray, (0,0,0), (0,h), (w,h), 3)
        pygame.draw.line(tray, (0,0,0), (w,0), (w,h), 3)  
        return tray
    
class sensor:
    
    def __init__(self, gm, pos, initCurView):
        self.pos = pos
        self.data = {}
        self.curCount = 0
        self.gm = gm
        self.initCurView = initCurView
        self.test = []
        
        x = pos[0]
        y = pos[1]
        xx = initCurView[0]
        yy = initCurView[1]
        posss = (xx+x, yy+y)
        self.whichWater = (posss[0]//8, posss[1]//8)
        self.kinds = list(self.gm.getWater(self.pos, self.initCurView).keys())
        for i in self.kinds:
            self.data[i] = []
        self.update()

    
    def update(self):
#         self.test.append(self.ourWater['Salinity'])
        row = self.gm.getWater(self.pos, self.initCurView)
        for i in self.kinds:
            self.data[i].append(row[i])
        for i in self.kinds:
            self.data[i] = self.data[i][-650:]
        
    def read(self, kind):
        if kind in list(self.data[0].keys()):
            d = self.data[kind][-650:]
            plt.plot(d)
            plt.ylabel(kind)
            plt.xlabel('time')
            return plt
        else:
            print("KIND NOT AVAILABLE")
```
