```python
from ipynb.fs.full.utils import util
from ipynb.fs.full.Cube import cube
import png
import pygame
from ipynb.fs.full.CritterModel import graphicsModel

class LokoGenerator:
    
    def __init__(self, surface, width, rows, pageManager, baseMap, miniMapName, lokoManager, clock):
        self.miniMapName = miniMapName
        self.fileName = baseMap
        self.baseMap = png.Reader("{}".format(baseMap))
        self.pixelMap = {(0, 6, 67): 'ocean2.png', (1, 18, 121): 'ocean1.png', (0, 77, 214): 'deep2.png' , (0, 82, 255): 'deep1.png',
           (0, 122, 170): 'shallows3.png', (0, 170, 214): 'water1.png', (0, 213, 255): 'water1.png', (1, 213, 255): 'shallows3.png', (255, 216, 255): 'sand3.png',
           (255, 213, 169): 'sand2.png', (255, 170, 121): 'sand1.png', (29, 38, 59): "reef.png", (1, 122, 170): "reef2.png", (0,65,0): "mangrove.png", (0, 131, 41): "grass.png", (135, 105, 0): "silt.png", 
            (133, 104, 0): "silt.png", (233, 251, 0): "tank.png", (192,192,192): "tank.png", (240, 213, 176): "tank.png", (254, 170, 121): 'sand1.png'}
        sand = ["sand1.png", "sand2.png", "sand3.png"]

        self.curvDelta = 5
        self.surface = surface
        self.width = width
        self.rows = rows
        self.win = self.surface
        self.clock = pygame.time.Clock()
        self.fps = [25]

        self.curView = (0,0)
        self.viewTrack = (0,0)
        self.gm = graphicsModel(440, 448, self.rows, miniMapName)
        self.mini = util.makeMaps(self.baseMap, self.pixelMap, self.gm, sand, self.width, self.rows)
        data = self.gm.viewport(self.curView, self.width, self.rows)
        self.fishArray = data[0]
        self.tideMakaha = data[1]
        self.tideFish = data[2]
        self.typeAdd = ""
        self.butToggle = False
#         self.tm = TaskManager(self.win)
        
        self.win.blit(self.mini, (0 , 0))
        
        self.exit = None
        self.pageManager = pageManager
        self.toRender = []
        self.toReturn = None
        self.homeBut = None
        
        self.lokoManager = lokoManager(self.surface, self.gm, self.width, self.rows, self.curView, self.fps)
        self.util = util()

#     def handleEvent(self, i):
#         pos = pygame.mouse.get_pos()
#         pass

    def handleEvent(self, i):
#         global mini, rows, typeAdd, miniMapHeight, butToggle, curView, taskStarter, toChoose, tm, curView

#         for i in pygame.event.get():
        pos = pygame.mouse.get_pos()
        if i.type == pygame.MOUSEBUTTONUP:
            if self.homeBut:
                if self.homeBut.collidepoint(pos):
                    self.is_running = False
                    self.toReturn = self.pageManager["Home"]

        if i.type == pygame.KEYDOWN:         
            if i.key == 1073741906: #up
                if self.curView[1] != 0:
                    self.curView = (self.curView[0], self.curView[1] - self.curvDelta)
#                     self.gm.viewport(self.curView, self.width, self.rows)
            if i.key == 1073741905: #down
                if self.curView[1] != 240:
                    self.curView =( self.curView[0], self.curView[1] + self.curvDelta)
#                     self.gm.viewport(self.curView, self.width, self.rows)
            if i.key == 1073741904: #left
                if self.curView[0] != 0:
                    self.curView =(self.curView[0] - self.curvDelta, self.curView[1])
#                     self.gm.viewport(self.curView, self.width, self.rows)
            if i.key == 1073741903: #right
                if self.curView[0] != 240:
                    self.curView = (self.curView[0] + self.curvDelta, self.curView[1])
#                     self.gm.viewport(self.curView, self.width, self.rows)
        self.lokoManager.handleEvent(i)

#             if i.type == pygame.MOUSEBUTTONUP:
#                 pos = pygame.mouse.get_pos()
#                 minz = self.mini.get_rect()
#                 if gameTracker["one"]:
#                     if taskStarter.collidepoint(pos):
#                         gameTracker["one"] = False
#                         gameTracker["two"] = True
#                 elif gameTracker["two"]:
#                     for i in toChoose:
#                         name = i["name"]
#                         region = i["region"]
#                         t = i["task"]
#                         if region.collidepoint((pos[0] - width/2 + width/4, pos[1] - width/2 + width/4)):
#                             print("SELECTED TASK: ", name)
#                             gameTracker["two"] = False
#                             gameTracker["three"] = True
#                             tm.startTask(t(width, rows, miniMapHeight, mini, gm, surface))
    
    def update(self):
        self.gm.update(self.curView)
        data = self.gm.viewport(self.curView, self.width, self.rows)
        self.fishArray = data[0]
        self.tideMakaha = data[1]
        self.tideFish = data[2]
        self.makahas = data[3]


    def reDraw(self):
        self.surface.fill((255,255,255))
        for i in self.gm.vp:
            i.draw(self.surface)
        for i in self.fishArray:
            i.draw(self.surface)
        if len(self.tideMakaha) != 0:
            tideStat = (self.gm.tideStatus[-1] - self.gm.tideStatus[-2])
            self.util.graphicsTide(self.tideMakaha, self.surface, tideStat)
        for i in self.tideFish:
            i.draw(self.surface)
        for i in self.makahas:
#             self, start, width, rows, color = (0, 0, 0), kind = 0, salinity = 35):
            
            cu = list(i.values())[0]
            nu = cube(list(i.keys())[0], self.width, self.rows, cu.color)
            nu.draw(self.surface)

            
        boxWid = 50
        moveView = ((self.curView[0])/2.6, (self.curView[1])/2.6)
        mini = util.makeMiniMap(self.rows, self.miniMapName)
        pygame.draw.line(mini, (255,255,255), moveView, (moveView[0], moveView[1] + boxWid))
        pygame.draw.line(mini, (255,255,255), moveView, (moveView[0]+boxWid, moveView[1]))
        pygame.draw.line(mini, (255,255,255), (moveView[0], moveView[1] + boxWid), (moveView[0]+ boxWid, moveView[1]+ boxWid))
        pygame.draw.line(mini, (255,255,255), (moveView[0] + boxWid, moveView[1]), (moveView[0] + boxWid, moveView[1] + boxWid))

        self.lokoManager.update()
        self.drawButtons()

        self.surface.blit(mini, (self.width - (55*3 + 10),self.width -  (56*3 + 10)))


#         if gameTracker["one"]:
#             taskSelectButton = begin()
#             surface.blit(taskSelectButton[0], (0,0))
#             taskStarter = taskSelectButton[1]
#         if gameTracker["two"]:
#             chooseTask = taskSelect()
#             surface.blit(chooseTask[0], (width/2 - width/4,width/2 - width/4))
#             toChoose = chooseTask[1]
#         if gameTracker["three"]:
#             curView = tm.update(curView)
#             pos = pygame.mouse.get_pos()
#             mPos = (round(pos[0]/5) - 1, round(pos[1]/5) - 1)
#             mPos = (mPos[0] + curView[0], mPos[1] + curView[1])
#             if mPos in list(gm.hoverable.keys()):
#                 surface.blit(sensorTooltip(mPos), (pos[0]- 36, pos[1] - 25))
    
    def drawButtons(self):
        homeBut = (pygame.image.load('images/home.png').convert_alpha())
        self.homeBut = self.surface.blit(homeBut, (50,50))
        
    def mainLoop(self):

        self.is_running = True
        count = 0
        while self.is_running:
            for event in pygame.event.get():
                # --- global events ---
                if event.type == pygame.QUIT:
                    self.is_running = False
                elif event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_ESCAPE:
                        self.is_running = False
                # --- objects events ---
                self.handleEvent(event)
            # --- updates ---
            self.update()
            # --- draws ---
            self.reDraw()

            pygame.display.update()
            count += 1
            # --- FPS ---
            self.clock.tick(self.fps[0])
            
        return self.toReturn
```
