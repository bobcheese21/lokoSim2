```python
import pygame
from ipynb.fs.full.Sensors import sensor, sensorDash
from ipynb.fs.full.utils import util
from ipynb.fs.full.Fish import SimpleFish

class heeiaManager:
    
    def __init__(self, surface, gm, width, rows, curView, fps):
        self.curvDelta = 5

        self.width = width
        self.rows = rows
        self.surface = surface
        self.gm = gm
        self.fps = fps
        self.curView = curView
        self.sensors = []
        self.sensorDash = sensorDash(self.sensors, self.width, self.surface)
        self.showSensorButton = False
        self.dashButton = pygame.Rect(0,0,0,0)
        self.unitRealm = pygame.Rect(0,0,0,0)
        self.butToggle = None
        
        self.decreaseSpeed = pygame.Rect(self.width - 55*3 - 190, self.width - 80, 44,50)
        self.increaseSpeed = pygame.Rect(self.width - 55*3 - 190 + 131., self.width - 80, 44,50)
        
        self.gm.fishArray[(100,100)] = SimpleFish((100,100))
        self.gm.fishArray[(150,150)] = SimpleFish((150,150))
        self.gm.fishArray[(200,200)] = SimpleFish((200,200))
        self.gm.fishArray[(250,250)] = SimpleFish((250,250))
        self.steps = {"twoClickMakaha": False}
        self.makahaTracker = []
        
    
    def handleEvent(self, i):
        if i.type == pygame.KEYDOWN:         
            if i.key == 1073741906: #up
                if self.curView[1] != 0:
                    self.curView = (self.curView[0], self.curView[1] - self.curvDelta)
#                     self.gm.viewport(self.curView, self.width, self.rows)
            if i.key == 1073741905: #down
                if self.curView[1] != 185:
                    self.curView =( self.curView[0], self.curView[1] + self.curvDelta)
#                     self.gm.viewport(self.curView, self.width, self.rows)
            if i.key == 1073741904: #left
                if self.curView[0] != 0:
                    self.curView =(self.curView[0] - self.curvDelta, self.curView[1])
#                     self.gm.viewport(self.curView, self.width, self.rows)
            if i.key == 1073741903: #right
                if self.curView[0] != 185:
                    self.curView = (self.curView[0] + self.curvDelta, self.curView[1])
#                     self.gm.viewport(self.curView, self.width, self.rows)
        if i.type == pygame.MOUSEBUTTONUP:
            pos = pygame.mouse.get_pos()
#             print("lokoManager curview", self.curView)
#             print("LOKO MANAGER FPS: ", self.fps)
            if self.increaseSpeed.collidepoint(pos):
                self.gm.timeFactor += .01
                self.gm.timeFactor = min(self.gm.timeFactor, 2)
            if self.decreaseSpeed.collidepoint(pos):
                self.gm.timeFactor -= .01
                self.gm.timeFactor = max(self.gm.timeFactor, .5)
            if self.dashButton.collidepoint(pos):
                    print("Dash clicked")
                    self.sensorDash.showing = True
            else:    
                poss = (round(pos[0]/5) - 1, round(pos[1]/5) - 1)
#                 print("POSS: ", poss)
#                 rBut = pygame.Rect(self.width - 57 + 5, self.width - self.miniMapHeight - self.unitSelect.get_rect().height - 20 + 10 , 40 , 40)
#                 mBut = pygame.Rect(self.width - 57 + 5, self.width - self.miniMapHeight - self.unitSelect.get_rect().height - 20 + 60, 40 , 40)
                mBut = pygame.Rect(self.width - 57 + 5, self.width - 55*3 + 10 - self.unitRealm.height - 40 + 10, 40 , 40)
#                 sBut = pygame.Rect(self.width - 57 + 5, self.width - 55*3 + 10 - self.unitRealm.height - 40 + 10, 40 , 40)
                sBut = pygame.Rect(self.width - 57 + 5, self.width - 55*3 + 10 - self.unitRealm.height - 40 + 60, 40 , 40)
#                 kBut = pygame.Rect(self.width - 57 + 5, self.width - self.miniMapHeight - self.unitSelect.get_rect().height - 20 + 160, 40 , 40)

                if mBut.collidepoint(pos):
                    self.gm.isTurbid = not self.gm.isTurbid
                    if self.butToggle == "MAKAHA":
                        self.butToggle = False
                    else:
                        self.butToggle = "MAKAHA"
#                                 self.cm.startScene(PlaceMakaha())
                        print("makaha selected")
#                 elif rBut.collidepoint(pos):
#                     if self.butToggle == "ROCK":
#                         self.butToggle = False
#                     else:
#                         self.butToggle = "ROCK"
#                         print("rockselect")
#                         self.cm.startScene(PlaceRocks())
                elif sBut.collidepoint(pos):
                    if self.butToggle == "SENSOR":
                        self.butToggle = False
                    else:
                        self.butToggle = "SENSOR"
                        print("s select")
#                         self.steps['sensor'] = True
#                 elif kBut.collidepoint(pos):
#                     if self.butToggle == "KAI":
#                         self.butToggle = False
#                     else:
#                         self.butToggle = "KAI"
#                         print("s select")
                else:
                    if self.butToggle == "MAKAHA":
                        if len(self.makahaTracker) < 2:
                            self.gm.click(poss,self.curView, self.butToggle)
                            self.makahaTracker.append({"poss": poss, "curView": self.curView})
                        if len(self.makahaTracker) == 2:
                            self.steps["twoClickMakaha"] = True
#                     if self.butToggle == "ROCK":
#                         self.rockTracker = self.rockTracker + 1
#                         self.gm.click(poss,curView, self.butToggle)
#                         if self.rockTracker > 10:
#                             print("ROCS IS TRUE")
#                             self.steps["rocks"] = Truese
                    if self.butToggle == "SENSOR":
                        self.sensors.append(sensor(self.gm, poss, self.curView))
                        self.gm.click(poss, self.curView, self.butToggle)
                    else:
                        self.gm.click(poss, self.curView, self.butToggle)

    def update(self):

        unitSelect = util.makeUnitSelect("Heeia", self.butToggle)
        self.unitRealm = self.surface.blit(unitSelect, (self.width - 57, self.width -  55*3 + 10 - unitSelect.get_rect().height - 40))
        if self.steps["twoClickMakaha"]:
            util.connectMakaha(self.makahaTracker, self.gm, self.steps)
        if len(self.sensors) > 0:
            tideDash = util.makeTideDash(self.gm.tideStatus)
            self.surface.blit(tideDash, (10,self.width - tideDash.get_rect().height - 10))
        alterSpeed = util.alterSpeed(self.gm.timeFactor)
        self.surface.blit(alterSpeed, (self.width-55*3 - 190,self.width-90))
        
class waikaluaManager:
    
    def __init__(self):
        pass
```
