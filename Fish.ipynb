```python
import pygame
import random as rand

class Fish:
    
    def __init__(self, surface, startingPos, curView, gm):
        self.swimDown = [pygame.image.load("graphicsPM/fishLeft.png"),pygame.image.load("graphicsPM/fishMid.png"),pygame.image.load("graphicsPM/fishRight.png"), pygame.image.load("graphicsPM/fishMid.png")]
        self.swimRight = [pygame.image.load("graphicsPM/fishSide1.png"),pygame.image.load("graphicsPM/fishSide4.png"),pygame.image.load("graphicsPM/fishSide2.png"),pygame.image.load("graphicsPM/fishSide1.png")]
        self.swimLeft = list(map(lambda x : pygame.transform.flip(x, True, False), self.swimRight))
        self.swimUp = list(map(lambda x : pygame.transform.flip(x, False, True), self.swimDown))
        self.surface = surface
        self.w = 25
        self.h = 25
        self.gm = gm
        self.di = ["down", "right", "left", "up"]
        self.curView = curView
        self.moveCount = 0
        self.curGMLocation = startingPos
        starter = rand.randint(0,3)
        collecter = [self.swimDown, self.swimRight, self.swimLeft, self.swimUp]
        self.surface.blit(collecter[starter][1], startingPos)
        self.curDir = self.di[starter]

    def update(self):
        if self.moveCount > 3:
            self.moveCount = 0
        direction = self.curDir
        if rand.randint(0,8) == 3:
            direction = self.di[rand.randint(0,3)]
        toTravel = 6
        if direction == "down":
            curGMLocation = (self.curGMLocation[0], self.curGMLocation[1] + toTravel)
        if direction == "right":
            curGMLocation = (self.curGMLocation[0] + toTravel, self.curGMLocation[1])
        if direction == "left":
            curGMLocation = (self.curGMLocation[0] - toTravel, self.curGMLocation[1])
        if direction == "up":
            curGMLocation = (self.curGMLocation[0], self.curGMLocation[1] - 3)
        
        poss = (round(curGMLocation[0]/5) - 1, round(curGMLocation[1]/5) - 1)
        cu = self.gm.gmCubeFromClick(poss, self.curView)
        if cu.kind != "WATER":
            print("AVOIDED")
            self.update()
        else:
            self.updateForced(direction, 3)
            self.curDir = direction
    
    def updateForced(self, direction, toTravel):
        if self.moveCount > 2:
            self.moveCount = 0
        if direction == "down":
            self.curGMLocation = (self.curGMLocation[0], self.curGMLocation[1] + toTravel)
            
            self.moveCount += 1
            self.surface.blit(self.swimDown[self.moveCount], self.curGMLocation)
        if direction == "right":
            self.curGMLocation = (self.curGMLocation[0] + toTravel, self.curGMLocation[1])
            
            self.moveCount += 1
            self.surface.blit(self.swimRight[self.moveCount], self.curGMLocation)
        if direction == "left":
            self.curGMLocation = (self.curGMLocation[0] - toTravel, self.curGMLocation[1])
            
            self.moveCount += 1
            self.surface.blit(self.swimLeft[self.moveCount], self.curGMLocation)
        if direction == "up":
            self.curGMLocation = (self.curGMLocation[0], self.curGMLocation[1] - toTravel)
            
            self.moveCount += 1
            self.surface.blit(self.swimUp[self.moveCount], self.curGMLocation)
        

class SimpleFish:
    
    def __init__(self , startingPos):
        self.swimDown = [pygame.image.load("graphicsPM/fishLeft.png"),pygame.image.load("graphicsPM/fishMid.png"),pygame.image.load("graphicsPM/fishRight.png"), pygame.image.load("graphicsPM/fishMid.png")]
        self.swimRight = [pygame.image.load("graphicsPM/fishSide1.png"),pygame.image.load("graphicsPM/fishSide4.png"),pygame.image.load("graphicsPM/fishSide2.png"),pygame.image.load("graphicsPM/fishSide1.png")]
        self.swimLeft = list(map(lambda x : pygame.transform.flip(x, True, False), self.swimRight))
        self.swimUp = list(map(lambda x : pygame.transform.flip(x, False, True), self.swimDown))
        self.pos = startingPos
        self.di = ["down", "right", "left", "up"]
        self.moveCount = 0
        
        starter = rand.randint(0,3)
        self.collector = {"down": self.swimDown, "right": self.swimRight, "left":  self.swimLeft,"up": self.swimUp}
        self.curDir = self.di[starter]
        self.toSwimTo = None
        self.renderPos = None
        self.debug = False
        self.secondaryPos = None
        
    def draw(self, surface):
        self.moveCount += 1
        if self.moveCount > 3:
            self.moveCount = 0

        imgs = self.collector[self.curDir]
        surface.blit(imgs[self.moveCount], (self.renderPos[0] * 5, self.renderPos[1]*5))
#                 poss = (round(pos[0]/5) - 1, round(pos[1]/5) - 1)

    def swimTo(self):
        if not self.toSwimTo:
            print("MUST SET FISH.toSwimTo")
# check if at toSwimTo, define behavior
        else:
            xGoal = self.toSwimTo[0]
            yGoal = self.toSwimTo[1]
            curX = self.pos[0]
            curY = self.pos[1]
            xDist = xGoal - curX
            yDist = yGoal - curY
            di = None
            xDelt = 0
            yDelt = 0
            if xDist > 0:
#                 travel to the right
                xDelt = 1
                di = "right"
            elif xDist < 0:
                xDelt = -1
                di = "left"
            if yDist > 0:
                if abs(yDist) > abs(xDist):
                    di = "down"
                # travel down 
                yDelt = 1
            elif yDist < 0:
                if abs(yDist) > abs(xDist):
                    di = "up"
                yDelt = -1

            self.curDir = di
            self.pos = (self.pos[0] + xDelt, self.pos[1] + yDelt)
        
class Mamo:
    
    def __init__(self , startingPos):
        self.swimDown = [pygame.image.load("graphicsPM/mamoLeft.png"),pygame.image.load("graphicsPM/mamoMid.png"),pygame.image.load("graphicsPM/mamoRight.png"), pygame.image.load("graphicsPM/mamoRight.png")]
        self.swimRight = [pygame.image.load("graphicsPM/mamoSide3.png"),pygame.image.load("graphicsPM/mamoSide2.png"),pygame.image.load("graphicsPM/mamoSide3.png"),pygame.image.load("graphicsPM/mamoSide1.png")]
        self.swimLeft = list(map(lambda x : pygame.transform.flip(x, True, False), self.swimRight))
        self.swimUp = list(map(lambda x : pygame.transform.flip(x, False, True), self.swimDown))
        self.pos = startingPos
        self.startingPos = startingPos
        self.di = ["down", "right", "left", "up"]
        idioDi = ["down", "right", "left", "up"]
        rand.shuffle(idioDi)
        self.idioDi = idioDi
        self.moveCount = 0
        self.counter = 0
        starter = rand.randint(0,3)
        self.collector = {"down": self.swimDown, "right": self.swimRight, "left":  self.swimLeft,"up": self.swimUp}
        self.curDir = self.di[starter]
        self.toSwimTo = None
        self.renderPos = None
        self.debug = False
        self.secondaryPos = None
        
    def draw(self, surface):
        imgs = self.collector[self.curDir]
        surface.blit(imgs[self.moveCount], (self.renderPos[0] * 5, self.renderPos[1]*5))
#                 poss = (round(pos[0]/5) - 1, round(pos[1]/5) - 1)

    def swimTo(self, tideFish):
        if not self.toSwimTo:
            print("MUST SET FISH.toSwimTo")
        if self.pos == self.toSwimTo:
            self.secondaryPos = self.pos
            self.secondary(tideFish)
        else:
            xGoal = self.toSwimTo[0]
            yGoal = self.toSwimTo[1]
            curX = self.pos[0]
            curY = self.pos[1]
            xDist = xGoal - curX
            yDist = yGoal - curY
            di = None
            xDelt = 0
            yDelt = 0
            if xDist > 0:
#                 travel to the right
                xDelt = 1
                di = "right"
            elif xDist < 0:
                xDelt = -1
                di = "left"
            if yDist > 0:
                if abs(yDist) > abs(xDist ):
                    di = "down"
                # travel down 
                yDelt = 1
            elif yDist < 0:
                if abs(yDist) > abs(xDist):
                    di = "up"
                yDelt = -1
            if di:  
                self.curDir = di
            self.pos = (self.pos[0] + xDelt, self.pos[1] + yDelt)
        self.moveCount += 1
        if self.moveCount > 3:
            self.moveCount = 0
    
    def secondary(self, locations):
        #swim in fashion without changing self.pos. Only change self.secondaryPos
        if self.pos != self.toSwimTo:
            self.pos = self.secondaryPos
            self.secondaryPos = None
            return self.pos
        else:
            self.moveCount += 1
            if self.moveCount > 3:
                self.moveCount = 0

            if rand.randint(0,2):
                self.counter += 1
                if self.counter > 3:
                    self.counter = 0
            toTravel = 1
            direction = self.idioDi[self.counter] 
            self.curDir = direction
            if direction == "down":
                curLoc = (self.secondaryPos[0], self.secondaryPos[1] + toTravel)
            if direction == "right":
                curLoc = (self.secondaryPos[0] + toTravel, self.secondaryPos[1])
            if direction == "left":
                curLoc = (self.secondaryPos[0] - toTravel, self.secondaryPos[1])
            if direction == "up":
                curLoc = (self.secondaryPos[0], self.secondaryPos[1] - toTravel)

            while curLoc in locations:
                self.counter += 1
                if self.counter > 3:
                    self.counter = 0
                    direction = self.idioDi[self.counter]
                    self.curDir = direction
                if direction == "down":
                    curLoc = (self.secondaryPos[0], self.secondaryPos[1] + toTravel)
                if direction == "right":
                    curLoc = (self.secondaryPos[0] + toTravel, self.secondaryPos[1])
                if direction == "left":
                    curLoc = (self.secondaryPos[0] - toTravel, self.secondaryPos[1])
                if direction == "up":
                    curLoc = (self.secondaryPos[0], self.secondaryPos[1] - toTravel)
            self.secondaryPos = curLoc
            return curLoc

    def swapLocs(self):
        if self.debug:
            print("preSwap pos: ", self.pos)
        temp = self.startingPos
        self.startingPos = self.toSwimTo
        self.toSwimTo = temp
        if self.debug:
            print("StartingPos: ", self.startingPos)
            print("toSwimTo: ", self.toSwimTo)
            print("pos: ", self.pos)
#     def update(self):
#         if self.moveCount > 3:
#             self.moveCount = 0
#         direction = self.curDir
#         if rand.randint(0,8) == 3:
#             direction = self.di[rand.randint(0,3)]
#         toTravel = 6
#         if direction == "down":
#             curGMLocation = (self.curGMLocation[0], self.curGMLocation[1] + toTravel)
#         if direction == "right":
#             curGMLocation = (self.curGMLocation[0] + toTravel, self.curGMLocation[1])
#         if direction == "left":
#             curGMLocation = (self.curGMLocation[0] - toTravel, self.curGMLocation[1])
#         if direction == "up":
#             curGMLocation = (self.curGMLocation[0], self.curGMLocation[1] - 3)
        
#         poss = (round(curGMLocation[0]/5) - 1, round(curGMLocation[1]/5) - 1)
#         cu = self.gm.gmCubeFromClick(poss, self.curView)
#         if cu.kind != "WATER":
#             print("AVOIDED")
#             self.update()
#         else:
#             self.updateForced(direction, 3)
#             self.curDir = direction
    
#     def updateForced(self, direction, toTravel):
#         if self.moveCount > 2:
#             self.moveCount = 0
#         if direction == "down":
#             self.curGMLocation = (self.curGMLocation[0], self.curGMLocation[1] + toTravel)
            
#             self.moveCount += 1
#             self.surface.blit(self.swimDown[self.moveCount], self.curGMLocation)
#         if direction == "right":
#             self.curGMLocation = (self.curGMLocation[0] + toTravel, self.curGMLocation[1])
            
#             self.moveCount += 1
#             self.surface.blit(self.swimRight[self.moveCount], self.curGMLocation)
#         if direction == "left":
#             self.curGMLocation = (self.curGMLocation[0] - toTravel, self.curGMLocation[1])
            
#             self.moveCount += 1
#             self.surface.blit(self.swimLeft[self.moveCount], self.curGMLocation)
#         if direction == "up":
#             self.curGMLocation = (self.curGMLocation[0], self.curGMLocation[1] - toTravel)
            
#             self.moveCount += 1
#             self.surface.blit(self.swimUp[self.moveCount], self.curGMLocation)
        
```
